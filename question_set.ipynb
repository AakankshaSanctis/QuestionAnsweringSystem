```python
import pandas as pd

data=pd.read_csv('WikiQA-train.tsv',delimiter='\t',encoding='utf-8')
data
```

```python
#Extracting the unique questions along with their questionID
def extract_questions(data):
    new_data=data.drop(['DocumentID','DocumentTitle','SentenceID','Sentence','Label'],axis=1)
    d=new_data.drop_duplicates()
    return d
```

```python
d=extract_questions(data)
d.to_csv('questions.csv',index=False)
```
